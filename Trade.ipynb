#### **Data Cleaning**
*So far we can note the following:*


*   The absence of NaN values(indicating no missing values are present).
* The absence of duplicated values.
*   The presence of complete zero values in many rows(thereby not adding significant value for the analysis). A possible reason for these zero values could be that data wasn't entered for those specific columns/rows. 
* Since the data primarly deals with import/export of various goods(for different countries). Dropping those rows of a column that contain zeroes seems to be more appropriate.
*   The column names(though comprehendable) are inconvenient(due to white spaces)(while attempting to draw the visualizations). 

* Presence of anomalies/outliers will be checked as we investigate further.

*In the next section we attempt to achieve the following:*
1. Remove the rows having zero values.
2. Altering the column names.

```
dataframe = dataset_raw_df

# Renaming the columns
dataframe.columns = [c.replace(' ', '_') for c in dataframe.columns]
dataframe.columns = [c.replace('-', '_') for c in dataframe.columns]
dataframe.columns = [c.replace('%', 'p') for c in dataframe.columns]
dataframe.columns = [c.replace('=', 'p') for c in dataframe.columns]
dataframe.columns = [c.replace('$', 'p') for c in dataframe.columns]
dataframe.columns = [c.replace('&', 'p') for c in dataframe.columns]
dataframe.columns = [c.replace('(', '_') for c in dataframe.columns]
dataframe.columns = [c.replace(')', '_') for c in dataframe.columns]
dataframe.columns = [c.replace('(', '_') for c in dataframe.columns]
dataframe.columns = [c.replace(',', '_') for c in dataframe.columns]


# Removing insignificant/zero values from dataset
# Removes rows having 60% of zero values
dataframe =  dataframe[(~dataframe.astype('bool')).mean(axis=1)<0.6]
# Removes columns having 60% of zero values
dataframe = dataframe.loc[:,(dataframe==0).mean() < 0.6]
dataframe

```

```
# Agricultural export percentage vs Ores & metal export percentage in United States over time using bubble charts.

data_1 = dataframe
Country = data_1.Country_Name == "United States"
data_1 = data_1[Country]
data_1 = data_1[data_1.Agricultural_raw_materials_exports__p_of_merchandise_exports_ != 0]
data_1 = data_1[data_1.Ores_and_metals_exports__p_of_merchandise_exports_ != 0]
data_1 = data_1[data_1.Export_value_index__2000_p_100_ != 0]

fig_1 = px.scatter(data_1, x="Agricultural_raw_materials_exports__p_of_merchandise_exports_",
                           y="Ores_and_metals_exports__p_of_merchandise_exports_",
                           size='Export_value_index__2000_p_100_', 
                           color="Year",
                           labels = {
                                    'Agricultural_raw_materials_exports__p_of_merchandise_exports_' : 'Agricultural raw material exports',
                                    'Ores_and_metals_exports__p_of_merchandise_exports_' : 'Ores & metal exports'
                                  },
                          title = "Agricultural export percentage vs Ores & metal export percentage in United States")
fig_1.show()
```

####**Terms:**
* Agricultural raw materials comprises of (crude materials except fuels) (crude fertilizers and minerals excluding coal, petroleum, and precious stones), and 28 (metalliferous ores and scrap).
* Ores and metals comprises of 27 (crude fertilizer, minerals nes); 28 (metalliferous ores, scrap); and 68 (non-ferrous metals).
* Export values are the ***current value of exports*** (f.o.b.) converted to US dollars and expressed as a percentage of the average for the base period.
* Current value of exports is the value for your export goods in currency.

####**Observations:**
* This bubble chart plots two variables i.e. Agricultural raw material exports vs Ore & metal exports for USA. The size of the bubble is indicated by the export value index.
* Over the years from 1980-2000 we see that there is a slight linear relationship between the two goods(agricultural raw materials and ores & metals) being exported. We see in those years agricutural raw materials exporting percentages were slightly higher than that of ores and metals.
* The years following 2000 an interesting observation can be made i.e. the exporting percentages of ore and metal have surpassed agricultural raw materials.
* The next aspect is the size of the bubbles, the export value index. One factor contributing to the economy is the export and import indices. In this graph we see that in USA the export value index was considerably small in 1980s to 1985. This implies that economy of USA during this period was weak. A small export value index could either mean that demand for these goods has decreased, but in the case of USA, the country had suffered a severe recession during this period of time resulting in decline of exports which in turn explains the small export value index.
* However over a period of time we see the economy of USA has improved(indicated by the increase in the export value index over time).

```
# Trends in High Technology, Computer Communication, Ores & metals export percentages in India using stream graphs.

plt.figure(figsize=(22,12))

data_6 = dataframe
country_6 = data_6.Country_Name == "India"
data_6 = data_6[country_6]
#data_6 = data_6.sort_values(by=['Year'])
data_6 = data_6[data_6.High_technology_exports__p_of_manufactured_exports_ != 0]
data_6 = data_6[data_6.Computer__communications_and_other_services__p_of_commercial_service_exports_ != 0]
data_6 = data_6[data_6.Ores_and_metals_exports__p_of_merchandise_exports_!= 0]
data_6 = data_6[['Ores_and_metals_exports__p_of_merchandise_exports_','High_technology_exports__p_of_manufactured_exports_','Computer__communications_and_other_services__p_of_commercial_service_exports_',]]
# blue - Ores and metal exports
# red - High technology exports
# violet - Computer and communication exports
data_6 = data_6.tail(25)
#data_6 (record starts from the year 1993)
data_6 = data_6.to_numpy()
fig, ax = plt.subplots()
ax.stackplot(range(25), data_6.T, baseline='wiggle')
plt.title("Trends in High Technology, Computer Communication, Ores & metals export percentages in India")
plt.show()
```

####**Observations:**
* Stream graphs are often used to show changes of different categories over time. The height of each individual stream shape shows how the value of that stream has changed over time. The length of the stream shape shows its duration. Colours have been used to differentiate between different categories.
* blue - Ores and metal exports,
  red - High technology exports,
  violet - Computer and communication exports
* From this we can see over time India's computer and communication export percentage has increased significantly(peaks in the graph). We can say that international telecommunications, the postal and courier serivices, computer data etc. provided by India are widely exported(indicating demand for these services from other countries).
* High technology goods include products with high R&D intensity, such as in aerospace, computers, pharmaceuticals, scientific instruments, and electrical machinery. Over time we can see a slight decrease in the export percentages of these goods suggesting that countries seek demand for these products from more developed countries(as exposure to high technology is a paramter in indicating the development of an economy).
* The product with a small export percentage in India goes to the ores and metals. We can see a decline in the export percentages over time.
